##Outbound & Inbound Flights

###Outbound Flights

```
%sql

DROP TABLE IF EXISTS group25.outbound_flights_main;

CREATE TABLE group25.outbound_flights_main AS (
SELECT
  ORIGIN,
  ORIGIN_AIRPORT_ID,
  FL_DATE,
  CAST(DEP_TIME / 100 AS INT) AS hour,
  COUNT(*) AS outbound_count
-- FROM group25.test4
FROM group25.airlines_main
WHERE DEP_TIME IS NOT NULL
GROUP BY 1, 2, 3, 4
ORDER BY 1, 2, 3, 4 ASC
);
```

```

%sql

SELECT * FROM group25.outbound_flights_main;
```

###Inbound Flights

```
%sql

DROP TABLE IF EXISTS group25.inbound_flights_main;

CREATE TABLE group25.inbound_flights_main AS (
  SELECT
    DEST,
    DEST_AIRPORT_ID,
    FL_DATE,
    CAST (ARR_TIME / 100 AS INT) AS HOUR,
    COUNT(*) AS inbound_count
--   FROM group25.test4
  FROM group25.airlines_main
  WHERE ARR_TIME IS NOT NULL
  GROUP BY 1, 2, 3, 4
  ORDER BY 1, 2, 3, 4 ASC
);
```

```
%sql

SELECT * FROM group25.inbound_flights_main;
```

###Join Inbound & Outbound

```
%sql

SELECT * FROM group25.inbound_outbound_flights_main;
```

```
%sql
DROP TABLE IF EXISTS group25.inbound_outbound_flights_test;

create table group25.inbound_outbound_flights_test as
select *, 
        to_timestamp(concat(cast(date as string),
                        '-', 
                        case when length(cast(hour as string)) = 1 then concat('0',cast(hour as string)) 
                              else cast(hour as string) end,
                        '-00-00'),
                        'yyyy-MM-dd-HH-mm-ss') as date_timestamp
                        from
(                        
select 
      case when a.origin is null then b.dest else a.origin end as Airport,
      case when a.fl_date is null then b.fl_date else a.fl_date end as date,
      case when a.hour is null then b.hour else a.hour end as hour,
      case when a.outbound_count is null then 0 else a.outbound_count end as outbound_count,
      case when b.inbound_count is null then 0 else b.inbound_count end as inbound_count
  from
    group25.outbound_flights_main a
       full outer join 
    group25.inbound_flights_main b
        on a.origin = b.dest
        and a.fl_date = b.fl_date
        and a.hour = b.hour)
```

```
%sql
select * from group25.inbound_outbound_flights_test LIMIT 10;
```

```
%sql
DROP TABLE IF EXISTS group25.inbound_outbound_flights;

create table group25.inbound_outbound_flights as (
  select *,
         to_date(date + INTERVAL 2 HOUR) as date_delta2h,
          hour(date + INTERVAL 2 HOUR) as hour_delta2h
  from group25.inbound_outbound_flights_test
);
```

```
%sql

SELECT * FROM group25.inbound_outbound_flights;
```

##Diverted Flights

We would like to know for a particular flight:
- At its origin airport, at that date, 2 hours before its departure time, how many flights has been diverted
- At its dest airport, at that date, 2 hours before its departure time, how many flights has been diverted

###Origin Airport

```
%sql

DROP TABLE IF EXISTS group25.outbound_flights_diverted_main;

CREATE TABLE group25.outbound_flights_diverted_main AS (
SELECT
  ORIGIN,
  ORIGIN_AIRPORT_ID,
  FL_DATE,
  CAST(DEP_TIME / 100 AS INT) AS hour,
  COUNT(*) AS diverted_count
-- FROM group25.PHASE_1_PROCESSED_AIRLINE
FROM group25.airlines_main
WHERE DIVERTED = 1 AND DEP_TIME IS NOT NULL
GROUP BY 1, 2, 3, 4
ORDER BY 1, 2, 3, 4 ASC
);
```

```
%sql

SELECT * FROM group25.outbound_flights_diverted_main;
```

###Destination Airport

```
%sql
-- Table: deverted_count
-- airport: DEST_AIRPORT_ID
-- date: FL_DATE
-- diverted_count: COUNT(rows)

DROP TABLE IF EXISTS group25.inbound_flights_diverted_main;

CREATE TABLE group25.inbound_flights_diverted_main AS (
  SELECT
    DEST,
    DEST_AIRPORT_ID, 
    FL_DATE,
    CAST(ARR_TIME / 100 AS INT) AS hour,
    COUNT(*) AS diverted_count
--   FROM group25.PHASE_1_PROCESSED_AIRLINE
  FROM group25.airlines_main
  WHERE DIVERTED = 1 AND ARR_TIME IS NOT NULL
  GROUP BY 1, 2, 3, 4
  ORDER BY 1, 2, 3, 4 ASC
);
```

```
%sql
SELECT * FROM group25.inbound_flights_diverted_main;
```

### Join diverted flights Tables

```
%sql
DROP TABLE IF EXISTS group25.inbound_outbound_diverted_flights_main_test;

create table group25.inbound_outbound_diverted_flights_main_test as
select *, 
        to_timestamp(concat(cast(date as string),
                        '-', 
                        case when length(cast(hour as string)) = 1 then concat('0',cast(hour as string)) 
                              else cast(hour as string) end,
                        '-00-00'),
                        'yyyy-MM-dd-HH-mm-ss') as date_timestamp
                        from
(                        
select 
      case when a.origin is null then b.dest else a.origin end as Airport,
      case when a.fl_date is null then b.fl_date else a.fl_date end as date,
      case when a.hour is null then b.hour else a.hour end as hour,
      case when a.diverted_count is null then 0 else a.diverted_count end as diverted_count_outbound,
      case when b.diverted_count is null then 0 else b.diverted_count end as diverted_count_inbound
  from
    group25.outbound_flights_diverted_main a
       full outer join 
    group25.inbound_flights_diverted_main b
        on a.origin = b.dest
        and a.fl_date = b.fl_date
        and a.hour = b.hour);
```

```
%sql
DROP TABLE IF EXISTS group25.inbound_outbound_diverted_flights_main;

create table group25.inbound_outbound_diverted_flights_main as (
  select *,
         to_date(date + INTERVAL 2 HOUR) as date_delta2h,
          hour(date + INTERVAL 2 HOUR) as hour_delta2h
  from group25.inbound_outbound_diverted_flights_main_test
);
```
